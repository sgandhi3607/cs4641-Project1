```python
# Helper function to discretize continuous data 
def discretize_data(input_dataframe, col_to_discretize, num_of_bins): 
    _ , bins = pd.qcut(input_dataframe[col_to_discretize], num_of_bins, retbins=True, 
                       labels=False, duplicates='drop')
    bin_min = bins[0]
    bin_max = bins[-1] 
    safety_width = (float(bin_max) - bin_min) / (num_of_bins / 2)
    bins = np.append([bin_min - safety_width], bins) 
    bins = np.append(bins, [bin_max + safety_width])
#     print "printing bins: "
#     print bins 
    
    input_dataframe[col_to_discretize] = pd.cut(input_dataframe[col_to_discretize], bins=bins, include_lowest=True)
    
    
```

```python
# Discretize using helper function 
#  'Education_Num', 'Capital_Gain', 'Capital_Loss', 'Hours/Week'
discretize_data(df, "Age", 10)
discretize_data(df, "fnlgwt", 10)
discretize_data(df, "Education_Num", 10)
discretize_data(df, "Capital_Gain", 5)
discretize_data(df, "Capital_Loss", 5)
discretize_data(df, "Hours/Week", 10)

# There may still be data points to left and right of all the bins, drop them
df.dropna(inplace=True)


print df.head()
```

```python
# Split into 30%  training data, 70% testing data
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.30, random_state=RANDOM_SEED)
```

```python
X_train = X_train[:100]
y_train = y_train[:100]


# continuous_columns = ['Age', 'fnlgwt', 'Education_Num', 'Capital_Gain', 'Capital_Loss', 'Hours/Week']
# continuous_df = X[continuous_columns]

X_train.drop(columns=['Age', 'fnlgwt', 'Education_Num', 'Capital_Gain', 'Capital_Loss', 'Hours/Week'], inplace=True)





print X_train.shape
print y_train.shape
```

```python
# Create lists to store results
train_error = list()
test_error = list()

# Begin training, varying max depth hyperparameter
for i in range(2, DT_DEPTH_LIMIT):
    clf = DecisionTreeClassifier(max_depth=i)
    clf.fit(X_train, y_train)

    train_error.append(mean_squared_error(y_train,
                                     clf.predict(X_train)))

    test_error.append(mean_squared_error(y_test,
                                     clf.predict(X_test)))
```
